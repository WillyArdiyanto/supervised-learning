```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
X = np.linspace(0,10,100)
y = np.sqrt(X) 
y1 = np.sin(X)
y2 = np.tan(X)
```

# Inisialisasi Figure dan Axes 

```python
fig = plt.figure()
ax = plt.axes()
```

```python
fig = plt.figure(figsize=(4,4))
ax = plt.axes()
ax.plot(X,y)
ax.plot(X,y1)
ax.plot(X,y2,'r+')
```

## Shortcut

```python
plt.figure(figsize=(4,4))
plt.plot(X,y1)
plt.plot(X,y2)
plt.xlim(0,10)
plt.ylim(-10,10)
plt.xlabel('X Axis')
plt.ylabel('y axis')
plt.title('Sin dan Tan');
```

# Scatter Plot 

```python
df = pd.read_csv('data/circle_data.csv', delimiter=";")
df.head()
df.info
```

```python
X = df.x
y = df.y
rad = df.radius
```

```python
plt.scatter(X,y)
plt.axis('equal')
```

```python
plt.scatter(X,y, c=rad,s=rad*50, cmap='seismic')
plt.axis('equal')
```

# text dan annotation 

```python
plt.figure()
numbers = np.random.randn(10000)
plt.hist(numbers,bins=25, density=1);
plt.ylim(0,1)
plt.axvline(0, color='r')
plt.text(0,0.8, 'mean=0', ha='center')
plt.annotate('mean=0', xy=(0,0.4), xytext=(1,0.5), ha='center', arrowprops={'arrowstyle':'->'})
```
