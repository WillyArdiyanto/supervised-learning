# Seaborn & Pandas Dataframe 

```python
df = pd.read_csv('data/cereal.csv', index_col='name')
df.head()
```

## Missing Value 

```python
plt.figure(figsize=(6,6))
sns.heatmap(df.isna(), cmap='plasma', cbar=True)
```

## Correlation Heatmap 

```python
plt.figure(figsize=(6,6))
sns.heatmap(df.corr(), cmap='plasma', vmin=-1, vmax=1, annot=True, fmt='.1f')
```

## Pairplot 

```python
sns.pairplot(df)
```

```python
sns.pairplot(df.dropna(), hue='shelf', vars=['calories', 'sugars', 'fat', 'rating'])
```

```python
sns.pairplot(df.dropna(), hue='shelf', vars=['calories', 'sugars', 'fat', 'rating'],kind='reg')
```

```python
sns.jointplot('potass','fiber',data=df, marginal_kws={'bins':25}, kind='reg')
```

# Categorical Plot


## Persebaran


```python
sns.catplot(x='mfr', y='rating', data=df)
```

```python
sns.boxplot(x='mfr', y='rating', data=df)
```

```python
sns.violinplot(x='mfr', y='rating', data=df)
```

## Counting 

```python
sns.catplot(x='mfr', data=df, kind='count')
```

```python
sns.countplot(x='mfr', data=df, hue='shelf')
```
